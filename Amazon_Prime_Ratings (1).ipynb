```python
import pandas as pd
```

```python
# Define the chunk size
chunk_size = 500000  # Adjust this number based on your needs

# Path to your file in Google Drive
file_path = '/content/drive/My Drive/Phase 5/meta_Movies_and_TV.jsonl'

# Read the first chunk of the JSON data
meta_chunk = pd.read_json(file_path, lines=True, chunksize=chunk_size)

# Get the first chunk
df_meta = next(meta_chunk)

# save chunk to a CSV file
df_meta.to_csv('/content/drive/My Drive/Phase 5/meta_movies_500000.csv', index=False)
```

```python
df_meta = pd.read_csv('/content/drive/My Drive/Phase 5/meta_movies_500000.csv')
df_meta.head(5)
```

```python
# prompt: I want to filter by Prime Video from the column main_category

# Filter the DataFrame by 'main_category'
prime_video_df = df_meta[df_meta['main_category'] == 'Prime Video']

# Display the filtered DataFrame
prime_video_df.head(5)
```

```python
# prompt: drop columns: subtitle, features,images,store,author,,bought_together,details

# Drop specified columns
columns_to_drop = ['subtitle', 'features', 'images', 'store', 'author', 'bought_together', 'details','videos']
prime_video_df = prime_video_df.drop(columns=columns_to_drop, errors='ignore')

prime_video_df.head(5)
```

```python
prime_video_df.info()
```

```python
# prompt: the category column is an object that looks like ['Comedy', 'Drama', 'Romance']	I want to pick the first word in that object. How can I achieve this?

prime_video_df['categories'] = prime_video_df['categories'].astype(str).str.split(',').str[0].str.strip()
prime_video_df.head()
```

```python
# prompt: How do I remove the characters [''] e.g ['Documentary']  ['Comedy'

prime_video_df['categories'] = prime_video_df['categories'].astype(str).str.replace(r"\['|'|\]", '', regex=True)
prime_video_df.head()
```

```python
prime_video_df.info()
```

```python
# Define the chunk size
chunk_size = 500000  # Adjust this number based on your needs

# Path to your file in Google Drive
file_path = '/content/drive/My Drive/Phase 5/Movies_and_TV.jsonl'

# Read the first chunk of the JSON data
reviews_chunk = pd.read_json(file_path, lines=True, chunksize=chunk_size)

# Get the first chunk
df_meta = next(reviews_chunk)

# save chunk to a CSV file
df_meta.to_csv('/content/drive/My Drive/Phase 5/reviews_movies_500000.csv', index=False)
```

```python
reviews_df = pd.read_csv('/content/drive/My Drive/Phase 5/reviews_movies_500000.csv')
reviews_df.head(5)
```

```python
# Drop specified columns
columns_to_drop = ['images','user_id','helpful_vote']
reviews_df = reviews_df.drop(columns=columns_to_drop, errors='ignore')

reviews_df.head(5)
```

```python
# prompt: I want to merge reviews_df with prime_video_df using the parent_asin column

# Merge the two DataFrames
merged_df = pd.merge(reviews_df, prime_video_df, on='parent_asin', how='inner')

# Display the merged DataFrame
merged_df.head()
```
